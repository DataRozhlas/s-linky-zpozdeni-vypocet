```python
import pandas as pd
import numpy as np
import json
```

```python
trains = pd.read_excel('./data/pid_linky.xls', sheetname='jednotlivé vlaky')
```

```python
zajmove = [
    'S9',
    'S1',
    'S2',
    'S7',
    'S4',
    #'S5',
    #'S3',
    #'S8',
    #'S6',
    #'S41'
          ]
```

```python
trains = trains[trains.Linka.isin(zajmove)]
```

```python
vlaky_linky = {}
for row in trains.values:
    trat = row[0]
    linka = row[1]
    usek = row[2]
    for vlak in row[3:]:
        if vlak == vlak:
            vlaky_linky[int(vlak)] = {'trat': trat, 'linka': linka, 'usek': usek}
            
with open('./data/vlaky_linky.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(vlaky_linky))
```

```python
trains.drop(['Trať', 'Linka', 'Úsek'],axis=1,inplace=True)
```

```python
trains = trains.applymap(lambda v: pd.to_numeric(v, errors='coerce'))
```

```python
tr_num = []
for row in trains.values:
    tr_num.extend(row)
```

```python
tr_num = np.array(tr_num)
```

```python
tr_num = tr_num[~np.isnan(tr_num)]
```

```python
tr_num = tr_num.astype('int')
```
